```python
from gammapy.datasets import FermiVelaRegion
from gammapy.morphology import Shell2D, Delta2D
from gammapy.irf import EnergyDependentTablePSF
from gammapy.image import coordinates

from astropy.units import Quantity
from astropy.coordinates import Angle
from astropy.modeling.models import Gaussian2D
from astropy.convolution.utils import discretize_model
from astropy.convolution import convolve
```

```python
# Get background model from Fermi diffuse model
energy_range = Quantity([1E4, 5E5], 'MeV')
fermi_diffuse = FermiVelaRegion.diffuse_model()
flux_background = fermi_diffuse.integral_flux_image(energy_range)

# Get PSF
fermi_psf = EnergyDependentTablePSF.from_fits(FermiVelaRegion.psf())

# Assume constant exposure
exposure = Quantity(0.1E8, 'cm2 s sr')

# Define source model
vela_junior = Shell2D(3E-6, -93.72, -1.24, 1, 0.5, normed=True)
vela_x = Gaussian2D(7.5E-7, -96.14, -3.09, 0.6, 0.6)

# Get coordinates from background reference 
glon, glat = coordinates(flux_background)
flux = Quantity(vela_x(glon, glat) + vela_junior(glon, glat), '1 / (cm2 s sr)')
```

```python
# Convolve with psf
pix_size = Angle(flux_background.header['cdelt1'], 'deg')
psf = fermi_psf.table_psf_in_energy_band(energy_range).kernel(pix_size)
flux_psf_convolved = convolve(flux, psf.value, boundary='wrap')
```

```python
flux_background.header['cdelt1'] = 0.125
```

```python
from aplpy import FITSFigure
from astropy.io import fits

textsize = 20
rc('xtick', labelsize=textsize)
rc('ytick', labelsize=textsize)
rc('font', size=textsize)
rc('text', usetex=True)
rc('figure', figsize=(8, 5))

# Compute counts image
counts_psf_convolved = np.random.poisson((flux_psf_convolved + flux_background.data.value) * exposure.value)
counts_hdu = fits.ImageHDU(counts_psf_convolved, flux_background.header)

f = FITSFigure(counts_hdu)
f.recenter(264, -1, width=18, height=12)
f.show_colorscale(vmin=0, vmax=20)
f.axis_labels.set_ytext('$\mathrm{Galactic\ Latitude\ [^{\circ}]}$')
f.tick_labels.set_yformat('ddd.')
f.axis_labels.set_xpad(10)
f.hide_xaxis_label()
f.tick_labels.set_xformat('ddd.')
f.add_colorbar()
f.colorbar.set_axis_label_text('$\mathrm{Counts}$')
f.axis_labels.set_xtext('$\mathrm{Galactic\ Longitude\ [^{\circ}]}$')
f.show_xaxis_label()
f.axis_labels.set_ypad(10)
f.save('counts.pdf')
```

```python
flux_hdu = fits.ImageHDU(flux.value + flux_background.data.value, flux_background.header)
f = FITSFigure(flux_hdu)
f.recenter(264, -1, width=18, height=12)
f.show_colorscale(vmin=1E-8, vmax=1.3E-6)
f.axis_labels.set_ytext('$\mathrm{Galactic\ Latitude\ [^{\circ}]}$')
f.tick_labels.set_yformat('ddd.')
f.axis_labels.set_xpad(10)
f.hide_xaxis_label()
f.tick_labels.set_xformat('ddd.')
f.add_colorbar()
ticks = np.linspace(1E-8, 1.3E-6, 10)
f.colorbar.set_ticks(ticks)
f.colorbar.set_axis_label_text('$\mathrm{Flux\ Density\ [cm^{-2} s^{-1} sr^{-1}]}$')
f.axis_labels.set_xtext('$\mathrm{Galactic\ Longitude\ [^{\circ}]}$')
f.show_xaxis_label()
f.axis_labels.set_ypad(10)
f.save('flux.pdf')
```
