```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
from astropy.units import Quantity
from astropy.io.fits import ImageHDU
from astropy.coordinates import Angle
from astropy.modeling.models import Gaussian2D
from astropy.convolution.utils import discretize_model
from astropy.convolution import convolve
from aplpy import FITSFigure
```

```python
from gammapy.datasets import FermiVelaRegion
from gammapy.morphology import Shell2D, Delta2D
from gammapy.irf import EnergyDependentTablePSF
from gammapy.image import coordinates
```

```python
# Get background model from Fermi diffuse model
energy_range = Quantity([10e3, 500e3], 'MeV')
# TODO: changing to the equivalent GeV here gives a different result (diffuse emission less bright) -> fix!!!
#energy_range = Quantity([10, 500], 'GeV')
fermi_diffuse = FermiVelaRegion.diffuse_model()
flux_background = fermi_diffuse.integral_flux_image(energy_range)

# Get PSF
fermi_psf = EnergyDependentTablePSF.from_fits(FermiVelaRegion.psf())

# Assume constant exposure
exposure = Quantity(0.1E8, 'cm2 s sr')

# Define source model
vela_junior = Shell2D(3E-6, -93.72, -1.24, 1, 0.5, normed=True)
vela_x = Gaussian2D(7.5E-7, -96.14, -3.09, 0.6, 0.6)

# Get coordinates from background reference 
glon, glat = coordinates(flux_background)
flux = Quantity(vela_x(glon, glat) + vela_junior(glon, glat), '1 / (cm2 s sr)')
```

```python
# Convolve with psf
pix_size = Angle(flux_background.header['cdelt1'], 'deg')
psf = fermi_psf.table_psf_in_energy_band(energy_range).kernel(pix_size)
flux_psf_convolved = convolve(flux, psf.value, boundary='wrap')
```

```python
# TODO: images below should be flipped left-right!!!
flux_background.header['cdelt1'] = 0.125
header = flux_background.header

# Compute flux and counts images
flux = flux.value + flux_background.data.value
counts = (flux_psf_convolved + flux_background.data.value) * exposure.value
counts = np.random.poisson(counts)
```

```python
def format_axes(fig):
    """Format figure axes labels and ticks nicely"""
    fig.axis_labels.set_xtext('Galactic longitude (deg)')
    fig.axis_labels.set_ytext('Galactic latitude (deg)')
    fig.tick_labels.set_xformat('ddd.')
    fig.tick_labels.set_yformat('ddd.')
```

```python
# Plot flux image
fig = FITSFigure(ImageHDU(1e6 * flux, header))
fig.recenter(264, -1, width=18, height=12)
fig.show_colorscale(vmin=1e-2, vmax=1.3e-0)
format_axes(fig)
fig.add_colorbar()
fig.colorbar.set_axis_label_text('$\mathrm{Flux\ Density\ [10^{-6} cm^{-2} s^{-1} sr^{-1}]}$')
fig.save('fermi_vela_model_flux.pdf')
```

```python
# Plot counts image
fig = FITSFigure(ImageHDU(counts, header))
fig.recenter(264, -1, width=18, height=12)
fig.show_colorscale(vmin=0, vmax=20)
format_axes(fig)
fig.add_colorbar()
fig.colorbar.set_axis_label_text('Counts per pixel')
fig.save('fermi_vela_model_counts.pdf')
```
