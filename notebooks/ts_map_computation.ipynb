# Computing TS maps with gammapy

```python
%matplotlib inline
from aplpy import FITSFigure
from astropy.io import fits
from astropy.convolution import Gaussian2DKernel
from gammapy.detect import compute_ts_map
```

```python
# Load data from files
maps = fits.open('../datasets/fermi_survey/all.fits.gz')

# Compute a source kernel in oversample mode
kernel = Gaussian2DKernel(2.5, mode='oversample')

# Compute a TS map with threshold=0, which skips the optimizing step,
# if TS < 0 for the initial flux estimate. 'ExpGammaMap' denotes to the exposure map.
result = compute_ts_map(maps['On'].data, maps['Background'].data,
                        maps['ExpGammaMap'].data, kernel)
print('TS map computation took {0:.2f} s'.format(result.runtime))
```

```python
# Plot TS map
f = FITSFigure(fits.ImageHDU(result.sqrt_ts, maps[0].header), figsize=(18, 4))
f.show_colorscale(vmin=0, vmax=10)
f.tick_labels.set_xformat('dd')
f.tick_labels.set_yformat('dd')
```

```python
# Plot flux map
f = FITSFigure(fits.ImageHDU(result.amplitude, maps[0].header), figsize=(18, 4))
f.show_colorscale(vmin=0, vmax=1E-9, stretch='sqrt')
f.tick_labels.set_xformat('dd')
f.tick_labels.set_yformat('dd')
```

```python
# Plot number of iterations
f = FITSFigure(fits.ImageHDU(result.niter, maps[0].header), figsize=(18, 4))
f.show_colorscale(vmin=0, vmax=5)
f.tick_labels.set_xformat('dd')
f.tick_labels.set_yformat('dd')
```
