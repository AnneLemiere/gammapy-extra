# Computing TS maps with gammapy

```python
from astropy.io import fits
from astropy.convolution import Gaussian2DKernel
from gammapy.detect import compute_ts_map
```

```python
# Load data from files
counts = fits.getdata('../datasets/source_diffuse_separation/galactic_simulations/fermi_counts.fits')
diffuse = fits.getdata('../datasets/source_diffuse_separation/galactic_simulations/fermi_diffuse.fits')[1:-1, :]
exposure = fits.getdata('../datasets/source_diffuse_separation/galactic_simulations/fermi_exposure_gal.fits')[0]

# Multiply with exposure to obtain a background counts image 
background = (diffuse * exposure)
```

```python
# Define a source kernel
kernel = Gaussian2DKernel(2.5, mode='oversample')

# Compute a TS map with debug=True and threshold=0
# 'debug=True' returns a TSMapResult object with ts, amplitude and niter
# 'threshold=0' skips the optimizing step, if TS < 0 for the initial flux estimate 
result = compute_ts_map(counts, background, exposure, kernel, threshold=0)
print 'TS map computation took {0:.2f} s'.format(result.runtime)
```

```python
# Plot TS
figsize(16, 10)
imshow(result.ts, interpolation='None', origin='lower', vmin=0, vmax=100)
colorbar(orientation='horizontal')
```

```python
# Plot flux
figsize(16, 10)
imshow(result.amplitude, interpolation='None', origin='lower', vmin=0, vmax=1E-9)
colorbar(orientation='horizontal')
```

```python
# Plot number of iterations
figsize(16, 10)
imshow(result.niter, interpolation='None', origin='lower', vmin=0, vmax=5)
colorbar(orientation='horizontal')
```
